```python
for x in range(1, 100):
    print ("page : {} ".format(x))
    payload['pageNumber'] = x
    print (payload)
    r = requests.post(url, data=payload)
    bs = BS(r.text)
    rt = bs.find('ul', class_='textFormat')
    if (rt == None):
        break
    rt2 = rt.findAll("li")
    for x in rt2:
        print ("-----------------")
        cnt = len(x.findAll("dd"))
        print (cnt)
        # 제목
        print (x.findAll("a")[1].text.strip())
        if (cnt >2):
            # 라벨 
            print (x.findAll("dd")[2].findAll("label"))
            # 라벨 값
            print (x.findAll("dd")[2].findAll("span"))
            # 카테고리
            if x.find("dd", class_='category') != None:
                print (" ".join(pattern.findall(x.find("dd", class_='category').text.strip())))
            else:
                print ("없어")
        else:
             # 라벨 
            print (x.findAll("dd")[1].findAll("label"))
            # 라벨 값
            print (x.findAll("dd")[1].findAll("span"))
            if x.find("dd", class_='category') != None:
                print (" ".join(pattern.findall(x.find("dd", class_='category').text.strip())))
            else:
                print ("없어")
```
